```python
x=np.linspace(1,10,100)
y=np.sin(x)
```

```python
plt.plot(x,y, '.-')
```

```python
plt.plot(x,y, ':')
```

```python
plt.plot(x, np.sin(x-0), color='blue', label='Sinx')
plt.plot(x, np.cos(x-0), color='red', label='Cosx')
plt.ylim(-1.5, 1.5)
plt.ylim(1, 15)
plt.grid()
plt.title("sin plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.axis('equal')
plt.legend()
plt.show()
```
